```python
df_temp['saleYear'] = df_temp.saledate.dt.year
df_temp['saleMonth'] = df_temp.saledate.dt.month
df_temp['saleDay'] = df_temp.saledate.dt.day
df_temp['saleDayofWeek'] = df_temp.saledate.dt.dayofweek
df_temp['saleDayofyear'] = df_temp.saledate.dt.dayofyear

#droping original sale date
df_temp.drop('saledate', axis= 1, inplace=True)
```

## Filling missing data

```python
for label, content in df_temp_own.items():
    if pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
# to check which numeric column have null value
for label, content in df_temp_own.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
for label,content in df_temp_own.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            df_temp_own[label + '_is_missing'] = pd.isnull(content)
            df_temp_own[label] = content.fillna(content.median())
```

```python
for label, content in df_temp_own.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

## Filling and turning categorical in numerical

```python
for label, content in df_temp_own.items():
    if not pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
for label,content in df_temp_own.items():
    if not pd.api.types.is_numeric_dtype(content):
        df_temp_own[label+'_is_missing'] = pd.isnull(content)
        df_temp_own[label] = pd.Categorical(content).codes+1
```
